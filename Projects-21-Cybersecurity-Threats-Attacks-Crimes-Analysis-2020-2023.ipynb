```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Membaca dataset
df = pd.read_csv(r"C:\Users\OPTION\Downloads\Python DA\Projects 21_Cybersecurity Threats, Attacks & Crimes Analysis (2020 - 2023) By Alfin Bahru Rahmika Umar\updated_cybersecurity_attacks.csv")
df
```

```python
# Menampilkan informasi dasar
print("Dimensi awal dataset:", df.shape)
```

```python
df.info()
df.describe()
```

```python
# Salin dataset asli untuk perbandingan
df_original = df.copy()

# 1. Deteksi dan hapus duplikat
print("Jumlah duplikat sebelum cleaning:", df.duplicated().sum())
df = df.drop_duplicates()
print("Jumlah duplikat setelah cleaning:", df.duplicated().sum())

# 2. Deteksi dan hapus missing values
print("Jumlah missing values per kolom sebelum cleaning:")
print(df.isnull().sum())
df = df.dropna()  # menghapus baris dengan missing values
print("Jumlah missing values per kolom setelah cleaning:")
print(df.isnull().sum())

# 3. Deteksi dan hapus outlier menggunakan metode IQR untuk beberapa kolom numerik
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

numeric_cols = ['Packet_Length', 'Anomaly_Scores', 'Source_Port', 'Destination_Port']
for col in numeric_cols:
    before = df.shape[0]
    df = remove_outliers(df, col)
    after = df.shape[0]
    print(f"Kolom {col}: Menghapus {before - after} baris outlier")

# Visualisasi: Histogram dan Box Plot untuk kolom Packet_Length sebagai contoh
plt.figure(figsize=(12, 5))
plt.subplot(1,2,1)
sns.histplot(df_original['Packet_Length'], color='blue', label='Original', kde=True, bins=30)
sns.histplot(df['Packet_Length'], color='orange', label='Cleaned', kde=True, bins=30)
plt.legend()
plt.title('Histogram Packet_Length')

plt.subplot(1,2,2)
sns.boxplot(data=[df_original['Packet_Length'], df['Packet_Length']], palette=['blue','orange'])
plt.xticks([0, 1], ['Original', 'Cleaned'])
plt.title('Box Plot Packet_Length')
plt.tight_layout()
plt.show()

# Perbandingan ukuran dataset
print("Ukuran dataset asli:", df_original.shape)
print("Ukuran dataset setelah cleaning:", df.shape)
```

```python
# Menyimpan dataset cleaned
cleaned_filename = 'updated_cybersecurity_attacks-cleaned.csv'
df.to_csv(cleaned_filename, index=False)
print(f"Dataset cleaned telah disimpan dengan nama: {cleaned_filename}")
```

```python
# Metrik
total_cases = df.shape[0]
total_users = df['User_Information'].nunique()
total_locations = df['Geo-location_Data'].nunique()

print(f"Total Kasus: {total_cases}")
print(f"Total User: {total_users}")
print(f"Total Lokasi: {total_locations}")
```

```python
# Pie Charts
fig, axs = plt.subplots(2, 3, figsize=(18, 12))
pie_features = {
    'Attack_Type': "Type of Cyberattacks",
    'Traffic_Type': "Types of Traffics",
    'Severity_Level': "Types of Levels",
    'Protocol': "Types of Protocols",
    'Browser': "Type of Browsers",
    'Log_Source': "Types Of Log Sources"
}

for ax, (col, title) in zip(axs.flatten(), pie_features.items()):
    counts = df[col].value_counts()
    ax.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    ax.set_title(title)
plt.tight_layout()
plt.show()
```

```python
# Bar Charts
plt.figure(figsize=(12,5))
# Device/OS distribution
sns.countplot(y='Device/OS', data=df, order=df['Device/OS'].value_counts().index)
plt.title('Distribution of Device/OS')
plt.show()
```

```python
# Top 10 Locations (Geo-location_Data) dari serangan
top_locations = df['Geo-location_Data'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_locations.values, y=top_locations.index, palette="viridis")
plt.xlabel('Jumlah Serangan')
plt.title('Top 10 Lokasi Cyberattacks')
plt.show()
```

```python
# Line Chart: Tahun vs Count of Attacks
# Grouping data berdasarkan Year dan Attack_Type serta menghitung jumlah serangan tiap kategori
yearly_attack_counts = df.groupby(['Year', 'Attack_Type']).size().reset_index(name='Count')

# Filter hanya untuk jenis serangan yang diinginkan
attack_types = ['Intrusion', 'Malware', 'DDoS']
yearly_attack_counts = yearly_attack_counts[yearly_attack_counts['Attack_Type'].isin(attack_types)]

plt.figure(figsize=(10,6))
sns.lineplot(data=yearly_attack_counts, x='Year', y='Count', hue='Attack_Type', marker='o')
plt.xlabel('Year')
plt.ylabel('Jumlah Serangan')
plt.title('Year vs Count of Attacks by Type')
plt.legend(title='Attack Type')
plt.show()
```
